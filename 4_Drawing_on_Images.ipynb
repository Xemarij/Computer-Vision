## **Drawing images and shapes using OpenCV**

Firsty, let's import OpenCV and numpy and define our imshow function

````
# Our Setup, Import Libaries, Create our Imshow Function and Download our Images
import cv2
import numpy as np
from matplotlib import pyplot as plt
````

Let's start off my making a blank square canvas

**Use Python numpy to create a black image.**

````
import numpy as np
import matplotlib.pyplot as plt

# Define the dimensions of the image
height = 480  # Height in pixels
width = 640   # Width in pixels
channels = 3  # Number of color channels (BGR or RGB)

# Create a black image (all pixel values set to 0)
black_image = np.zeros((height, width, channels), dtype=np.uint8)

# Display the black image using matplotlib
plt.imshow(black_image)
plt.title('Black Image')
plt.axis('off')  # Hide axes
plt.show()
````

### **Let's draw a line over our black square**

```cv2.line(image, starting cordinates, ending cordinates, color, thickness)```

**Use the same image and draw a diagonal blue line over the it using Python cv2.**

````
# Define the start and end points for the diagonal line
start_point = (0, 0)           # Top-left corner
end_point = (width, height)    # Bottom-right corner

# Define the color in BGR (Blue color)
color = (255, 0, 0)  # BGR format (full blue)

# Define the thickness of the line
thickness = 5

# Draw the diagonal line on the image
cv2.line(black_image, start_point, end_point, color, thickness)

# Display the image with the diagonal blue line using matplotlib
plt.imshow(cv2.cvtColor(black_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title('Black Image with Diagonal Blue Line')
plt.axis('off')  # Hide axes
plt.show()

# Note this is an inplace operation, meaning it changes the input image
# Unlike many other OpenCV functions that return a new image leaving the input unaffected
# Remember our image was the black canvas
````

### **Now let's draw some cirlcles?**

```cv2.cirlce(image, center, radius, color, fill)```

**Give me Python cv2 code to draw a green circle on the black_image.**

````
# Create a black image (all pixel values set to 0)
black_image = np.zeros((height, width, channels), dtype=np.uint8)

# Define the center and radius of the circle
center = (width // 2, height // 2)  # Center of the image
radius = 100                         # Radius of the circle

# Define the color in BGR (Green color)
green_color = (0, 255, 0)  # BGR format for green

# Define the thickness of the circle outline (use -1 to fill the circle)
thickness = 5  # Positive value for outline, -1 for filled circle

# Draw the green circle on the image
cv2.circle(black_image, center, radius, green_color, thickness)

# Display the image with the green circle using matplotlib
plt.imshow(cv2.cvtColor(black_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title('Black Image with Green Circle')
plt.axis('off')  # Hide axes
plt.show()
````

**Now fill the circle.**

````
# Define the dimensions of the image
height = 480  # Height in pixels
width = 640   # Width in pixels
channels = 3  # Number of color channels (BGR)

# Create a black image (all pixel values set to 0)
black_image = np.zeros((height, width, channels), dtype=np.uint8)

# Define the center and radius of the circle
center = (width // 2, height // 2)  # Center of the image
radius = 100                         # Radius of the circle

# Define the color in BGR (Green color)
green_color = (0, 255, 0)  # BGR format for green

# Define the thickness of the circle (use -1 to fill the circle)
thickness = -1  # Set to -1 for a filled circle

# Draw the filled green circle on the image
cv2.circle(black_image, center, radius, green_color, thickness)

# Display the image with the filled green circle using matplotlib
plt.imshow(cv2.cvtColor(black_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title('Black Image with Filled Green Circle')
plt.axis('off')  # Hide axes
plt.show()
````

**Make the circle partially filled.**

````
# Define the dimensions of the image
height = 480  # Height in pixels
width = 640   # Width in pixels
channels = 3  # Number of color channels (BGR)

# Create a black image (all pixel values set to 0)
black_image = np.zeros((height, width, channels), dtype=np.uint8)

# Define the center and radius of the circle
center = (width // 2, height // 2)  # Center of the image
radius = 100                         # Radius of the circle

# Define the color in BGR (Green color)
green_color = (0, 255, 0)  # BGR format for green
background_color = (0, 0, 0)  # Black background

# Draw the filled green circle
cv2.circle(black_image, center, radius, green_color, -1)

# Draw a smaller black circle on top to create the illusion of a partial fill
inner_radius = 60  # Smaller radius for the inner circle
cv2.circle(black_image, center, inner_radius, background_color, -1)

# Display the image with the partially filled green circle using matplotlib
plt.imshow(cv2.cvtColor(black_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title('Black Image with Partially Filled Green Circle')
plt.axis('off')  # Hide axes
plt.show()
````

### **Polygons**

```cv2.polylines(image, points, Closed?, color, thickness)```

if Closed = True, we join the first and last points.

**Create a red polygon on the black_image.**

````
# Define the dimensions of the image
height = 480  # Height in pixels
width = 640   # Width in pixels
channels = 3  # Number of color channels (BGR)

# Create a black image (all pixel values set to 0)
black_image = np.zeros((height, width, channels), dtype=np.uint8)

# Define the vertices of the polygon (e.g., a triangle)
vertices = np.array([[200, 100], [300, 300], [100, 300]], np.int32)  # Triangle vertices
vertices = vertices.reshape((-1, 1, 2))  # Reshape for polylines function

# Define the color in BGR (Red color)
red_color = (0, 0, 255)  # BGR format for red

# Draw the polygon outline
cv2.polylines(black_image, [vertices], isClosed=True, color=red_color, thickness=3)

# Alternatively, to fill the polygon, you can use fillPoly
# cv2.fillPoly(black_image, [vertices], red_color)

# Display the image with the red polygon using matplotlib
plt.imshow(cv2.cvtColor(black_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title('Black Image with Red Polygon')
plt.axis('off')  # Hide axes
plt.show()
````

**Create a Trapezoid.**

````
# Define the dimensions of the image
height = 480  # Height in pixels
width = 640   # Width in pixels
channels = 3  # Number of color channels (BGR)

# Create a black image (all pixel values set to 0)
black_image = np.zeros((height, width, channels), dtype=np.uint8)

# Define the vertices of the quadrilateral (trapezoid)
vertices = np.array([[200, 100], [400, 100], [350, 300], [250, 300]], np.int32)  # Trapezoid vertices
vertices = vertices.reshape((-1, 1, 2))  # Reshape for polylines function

# Define the color in BGR (Red color)
red_color = (0, 0, 255)  # BGR format for red

# Draw the polygon outline
cv2.polylines(black_image, [vertices], isClosed=True, color=red_color, thickness=3)

# Alternatively, to fill the polygon, you can use fillPoly
# cv2.fillPoly(black_image, [vertices], red_color)

# Display the image with the quadrilateral using matplotlib
plt.imshow(cv2.cvtColor(black_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.title('Black Image with Quadrilateral (Trapezoid)')
plt.axis('off')  # Hide axes
plt.show()
````

### **And now for adding text with cv2.putText**

```cv2.putText(image, 'Text to Display', bottom left starting point, Font, Font Size, Color, Thickness)```

**Available Fonts**
- FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN
- FONT_HERSHEY_DUPLEX,FONT_HERSHEY_COMPLEX
- FONT_HERSHEY_TRIPLEX, FONT_HERSHEY_COMPLEX_SMALL
- FONT_HERSHEY_SCRIPT_SIMPLEX
- FONT_HERSHEY_SCRIPT_COMPLEX

````
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black canvas (500x500 pixels)
canvas = np.zeros((500, 500, 3), dtype="uint8")

# Set text parameters
text = "Hello World!"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1.5
color = (255, 255, 255)  # White text
thickness = 2
position = (50, 250)  # Position where the text will appear

# Write the text on the canvas
cv2.putText(canvas, text, position, font, font_scale, color, thickness)

# Convert the image from BGR (OpenCV format) to RGB (Matplotlib format)
canvas_rgb = cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(canvas_rgb)
plt.axis('off')  # Hide the axes
plt.show()
````
