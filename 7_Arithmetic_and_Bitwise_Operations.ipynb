## **Arithmetic Operations**

These are simple operations that allow us to directly add or subract to the color intensity.

Calculates the per-element operation of two arrays. The overall effect is increasing or decreasing brightness.

**Read an image with Python cv2 as a grayscale image. Create a numpy matrix equation 'M' that will multiply all the grayscale image with all 100 values to increase the brightness of the image. Then use matplotlib to show two images and also show their sizes.**

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('images/4.jpg', cv2.IMREAD_GRAYSCALE)

# Display original image size
original_size = image.shape

# Define the brightness increase matrix 'M' (we'll add 100 to each pixel)
M = np.ones(image.shape, dtype="uint8") * 100

# Increase brightness by adding matrix M to the image, and clip values to [0, 255] range
bright_image = cv2.add(image, M)

# Display new image size (same as original, since dimensions donâ€™t change)
bright_size = bright_image.shape

# Plot both images with their sizes using matplotlib
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title(f'Original Image\nSize: {original_size}')
plt.axis('off')

# Brightened image
plt.subplot(1, 2, 2)
plt.imshow(bright_image, cmap='gray')
plt.title(f'Brightened Image\nSize: {bright_size}')
plt.axis('off')

# Show the plots
plt.show()
```

#### **Decreasing Brightness**

**Now subtract the M from the image to decrease the brightness and print the images with matplotlib.**

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('./images/4.jpg', cv2.IMREAD_GRAYSCALE)

# Define the brightness decrease matrix 'M' (we'll subtract 100 from each pixel)
M = np.ones(image.shape, dtype="uint8") * 100

# Decrease brightness by subtracting matrix M from the image, and clip values to [0, 255] range
dark_image = cv2.subtract(image, M)

# Display both original and darkened images with their sizes using matplotlib
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Darkened image
plt.subplot(1, 2, 2)
plt.imshow(dark_image, cmap='gray')
plt.title('Darkened Image')
plt.axis('off')

# Show the plots
plt.show()
```

## **Bitwise Operations and Masking (AND, OR, XOR, NOT)**

To demonstrate these operations let's create some simple images

**Use Python cv2 code to craete two black images. On top of one image create a white square and on top of the other back image create an ellipse. Use these two images and perform bitwise And, OR, XOR, and NOT functions on them. At the end print all the images with matplotlib.**

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create two black images
image1 = np.zeros((300, 300), dtype="uint8")
image2 = np.zeros((300, 300), dtype="uint8")

# Draw a white square on the first image
cv2.rectangle(image1, (50, 50), (250, 250), 255, -1)

# Draw a white ellipse on the second image
cv2.ellipse(image2, (150, 150), (150, 100), 0, 0, 360, 255, -1)

# Perform bitwise operations
bitwise_and = cv2.bitwise_and(image1, image2)
bitwise_or = cv2.bitwise_or(image1, image2)
bitwise_xor = cv2.bitwise_xor(image1, image2)
bitwise_not1 = cv2.bitwise_not(image1)
bitwise_not2 = cv2.bitwise_not(image2)

# Display the images
plt.figure(figsize=(12, 8))

# Original images
plt.subplot(2, 3, 1)
plt.imshow(image1, cmap='gray')
plt.title('Image 1 (Square)')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(image2, cmap='gray')
plt.title('Image 2 (Ellipse)')
plt.axis('off')

# Bitwise AND
plt.subplot(2, 3, 3)
plt.imshow(bitwise_and, cmap='gray')
plt.title('Bitwise AND')
plt.axis('off')

# Bitwise OR
plt.subplot(2, 3, 4)
plt.imshow(bitwise_or, cmap='gray')
plt.title('Bitwise OR')
plt.axis('off')

# Bitwise XOR
plt.subplot(2, 3, 5)
plt.imshow(bitwise_xor, cmap='gray')
plt.title('Bitwise XOR')
plt.axis('off')

# Bitwise NOT on Image 1
plt.subplot(2, 3, 6)
plt.imshow(bitwise_not1, cmap='gray')
plt.title('Bitwise NOT (Image 1)')
plt.axis('off')

# Show the plots
plt.show()
```
