# **Color Spaces**

In this lesson we'll learn to:
1. View the individual channels of an RGB Image
2. Manipulate a color space

```
# Our Setup, Import Libaries, Create our Imshow Function and Download our Images
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

**One line python cv2 code to load the image.**

```
# Load our input image
image = cv2.imread('./images/3.jpg')
```

**Python matplotlib code to display the loaded image.**

```
# Convert the image from BGR (OpenCV default) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide axes for a cleaner display
plt.show()
```

```

image = cv2.imread('./images/4.jpg')

# Split the image into B, G, R channels
B, G, R = cv2.split(image)

# Create a figure to display all channels
plt.figure(figsize=(10, 10))


# Display Blue channel
plt.subplot(1, 3, 1)
plt.imshow(B)
plt.title('Blue Channel')
plt.axis('off')

# Display Green channel
plt.subplot(1, 3, 2)
plt.imshow(G)
plt.title('Green Channel')
plt.axis('off')

# Display Red channel
plt.subplot(1, 3, 3)
plt.imshow(R)
plt.title('Red Channel')
plt.axis('off')
```

```
# Use cv2.split to get each color space separately

print(B.shape)
print(G.shape)
print(R.shape)
```

**Give Python cv2 code to merge the red, green and blue channels.**

```

# Merge the channels back into a single image
merged_image = cv2.merge([B, G, R])

# Display the original and merged images
cv2.imshow('Original Image', image)
cv2.imshow('Merged Image', merged_image)
```

```
# Convert the merged image from BGR to RGB format for displaying
merged_image_rgb = cv2.cvtColor(merged_image, cv2.COLOR_BGR2RGB)

# Display the original and merged images using matplotlib
plt.figure(figsize=(10, 5))

# Display the merged image
plt.subplot(1, 2, 2)
plt.imshow(merged_image_rgb)
plt.title('Merged Image')
plt.axis('off')

# Show the figure
plt.tight_layout()
plt.show()
```

```
# Load the image
image = cv2.imread('./images/5.jpg')

# Split the image into B, G, R channels
B, G, R = cv2.split(image)

# Amplify the blue channel
B_amplified = cv2.multiply(B, 2)  # Multiply the blue channel by 2

# Clip the values to ensure they are within the valid range [0, 255]
B_amplified = np.clip(B_amplified, 0, 255).astype(np.uint8)

# Merge the channels back into a single image
merged_image = cv2.merge([B_amplified, G, R])

# Convert the merged image from BGR to RGB format for displaying
merged_image_rgb = cv2.cvtColor(merged_image, cv2.COLOR_BGR2RGB)

# Display the original and modified images using matplotlib
plt.figure(figsize=(10, 5))

# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Display the modified image
plt.subplot(1, 2, 2)
plt.imshow(merged_image_rgb)
plt.title('Amplified Blue Channel')
plt.axis('off')

# Show the figure
plt.tight_layout()
plt.show()
```

## **The HSV Color Space**

![](https://upload.wikimedia.org/wikipedia/commons/f/f2/HSV_color_solid_cone.png)

- Hue: 0 - 179
- Saturation: 0 - 255
- Value (Intensity): 0 - 255

```
# Load the image
image = cv2.imread('./images/5.jpg')

# Convert the image from BGR to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Display the original and HSV images using matplotlib
plt.figure(figsize=(10, 5))

# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert to RGB for correct color display
plt.title('Original Image (BGR)')
plt.axis('off')

# Display the HSV image
plt.subplot(1, 2, 2)
plt.imshow(hsv_image)  # HSV values will not display correctly as colors
plt.title('HSV Image')
plt.axis('off')

# Show the figure
plt.tight_layout()
plt.show()
```

#### This does not look right....why?

Because our plotting function was designed for RGB only images, not HSV

### **Let's view each channel type in the HSV Color Space representation**

**Use matplotlib to display the each channel in HSV separately.**

```
# Load the image
image = cv2.imread('./images/5.jpg')

# Convert the image from BGR to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Split the HSV image into H, S, and V channels
H, S, V = cv2.split(hsv_image)

# Display the original image and HSV channels using matplotlib
plt.figure(figsize=(12, 6))

# Display the original image
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert to RGB for correct color display
plt.title('Original Image (BGR)')
plt.axis('off')

# Display the Hue channel
plt.subplot(2, 2, 2)
plt.imshow(H)  # Display Hue channel in grayscale
plt.title('Hue Channel')
plt.axis('off')

# Display the Saturation channel
plt.subplot(2, 2, 3)
plt.imshow(S)  # Display Saturation channel in grayscale
plt.title('Saturation Channel')
plt.axis('off')

# Display the Value channel
plt.subplot(2, 2, 4)
plt.imshow(V)  # Display Value channel in grayscale
plt.title('Value Channel')
plt.axis('off')

# Show the figure
plt.tight_layout()
plt.show()
```
