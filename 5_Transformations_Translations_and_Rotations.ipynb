# **Transformations - Translations and Rotations**

In this lesson we'll learn to:
1. Perform Image Translations
2. Rotations with getRotationMatrix2D
3. Rotations with Transpose
4. Flipping Images

````
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image (replace 'image.jpg' with your image file)
image = cv2.imread('./images/2.jpg')

# Get image dimensions
height, width = image.shape[:2]

# Define the translation matrix
# Quarter of the height down and quarter of the width to the right
tx = width // 4
ty = height // 4
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])

# Perform the translation
translated_image = cv2.warpAffine(image, translation_matrix, (width, height))

# Convert the image from BGR (OpenCV format) to RGB (Matplotlib format)
translated_image_rgb = cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB)

# Display the translated image using matplotlib
plt.imshow(translated_image_rgb)
plt.axis('off')  # Hide the axes
plt.show()
````

### **Translations**

This an affine transform that simply shifts the position of an image. (left or right).

### **Rotations**



**Write a python cv2 code to rotate an image 90 degree counter clockwise.**

````
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('./images/4.jpg')

# Rotate the image 90 degrees counterclockwise
rotated_image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Convert the image from BGR (OpenCV format) to RGB (Matplotlib format)
rotated_image_rgb = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)

# Display the rotated image using matplotlib
plt.imshow(rotated_image_rgb)
plt.axis('off')  # Hide the axes
plt.show()
````

# To Rotate the Image within the Canvas use the "Rotation Matrix function".
Prompt: **Use roation matrix function to rotate the image 90 degrees counterclockwise**

````
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('./images/4.jpg')

# Get the dimensions of the image
(h, w) = image.shape[:2]

# Define the center of rotation as the center of the image
center = (w // 2, h // 2)

# Define the rotation matrix for a 90-degree counterclockwise rotation
rotation_matrix = cv2.getRotationMatrix2D(center, 90, 1.0)

# Perform the rotation using warpAffine
rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))

# Convert the image from BGR (OpenCV format) to RGB (Matplotlib format)
rotated_image_rgb = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)


# Display the rotated image using matplotlib
plt.imshow(rotated_image_rgb)
plt.axis('off')  # Hide the axes
plt.show()
````

Notice all the black space surrounding the image.

We could now crop the image as we can calculate it's new size (we haven't learned cropping yet!).

### **Rotations with cv2.transpose** (less flexible)




**Use cv2.transpose to rotate an image counterclockwise to 90 degrees and display the original and transponsed image with Matplotlib.**

````
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('./images/4.jpg')

# Convert the image from BGR (OpenCV format) to RGB (Matplotlib format)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Transpose the image
transposed_img = cv2.transpose(img)

# Display the original and transposed images
plt.imshow(img)
plt.axis('off')  # Hide the axes
plt.show()

plt.imshow(transposed_img)
plt.axis('off')  # Hide the axes
plt.show()
````

# Horizonatlly Flip an Image

````
# Load the image
image = cv2.imread('./images/4.jpg')  # Replace with your image path

# Flip the image horizontally
flipped_image = cv2.flip(image, 1)

# Convert the color from BGR to RGB for Matplotlib display
flipped_image_rgb = cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB)

# Display the flipped image using Matplotlib
plt.imshow(flipped_image_rgb)
plt.axis('off')  # Hide axis
plt.show()
````
