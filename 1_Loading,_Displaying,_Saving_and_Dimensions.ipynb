### **Loading Images**

```
# Load an image using 'imread' specifying the path to image
image = cv2.imread('./images/1.jpg')
```

### **Displaying Images**

Prompt:**Use Python code to display an image using matplotlib**

```
import matplotlib.pyplot as plt
# Convert the image from BGR (OpenCV format) to RGB (Matplotlib format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using Matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide axis
plt.show()
```

Let's create a simple function to make displaying our images simpler and easier

Prompt:**Create me a function to display images Using Python Matplotlib**

```
def display_image(image):
    image = cv2.imread('./images/3.jpg')
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Hide the axis
    plt.show()
```

```
# Let's test it out
display_image('3.jpg')
```

### **Displaying Image Dimensions**

Remember Images are arrays:

We need to use numpy to perform this operation. Don't worry, numpy will become one of your best friends if you're learning a Data Science and Computer Vision.

Prompt: **Give me Python numpy one line code to get image dimensions**

```
# Import numpy
import numpy as np

# Get dimensions in one line
height, width, channels = cv2.imread('./images/1.jpg').shape
print(f"Height: {height}, Width: {width}, Channels: {channels}")
```

```
image = cv2.imread('./images/1.jpg')
height = image.shape[0]  # Height
print(height)
```

We can also store the image in a variable and we call the variable 'img'

Prompt: **Store the above image dimensions in a variable called "img"**

```
# Store the dimensions in the variable 'img'
img = cv2.imread('./images/1.jpg').shape

# Unpack the dimensions if needed
height, width, channels = img

print(f"Height: {height}, Width: {width}, Channels: {channels}")
```

```
# To access a dimension, simply index it by using 0, 1 or 2.
img = cv2.imread('./images/1.jpg').shape
# Access and print the dimensions
height = img[0]   # Height of the image
width = img[1]    # Width of the image
channels = img[2]  # Number of color channels

print(f"Height: {height}, Width: {width}, Channels: {channels}")
```

You can see the first dimension is the height and it's 960 pixels

The second dimension is the width which is 1280 pixels.

We can print them out nicely like this:

```
print('Height of Image: {} pixels'.format(int(img[0])))
print('Width of Image: {} pixels'.format(int(img[1])))
print('Depth of Image: {} colors components'.format(int(img[2])))
```
